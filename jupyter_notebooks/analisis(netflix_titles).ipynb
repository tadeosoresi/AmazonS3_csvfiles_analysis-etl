```python
### Vamos a empezar con logica de los datos (fundamentalmente variables cuantitativas)
### Year no logicos
import numpy as np
def parse_year(year):
    try:
        year = int(year)
        return year
    except (TypeError, ValueError):
        return None
df_copy['release_year'] = df_copy['release_year'].apply(parse_year)
```

```python
result['duration'] = result['duration'].apply(lambda duration: int(re.findall(r'\d+', duration)[0]))
```
