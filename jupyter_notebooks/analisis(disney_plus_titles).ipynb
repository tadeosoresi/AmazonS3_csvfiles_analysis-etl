```python
### Estan en formato string, de los cuales tenemos muchos paises en un mismo string separados por coma.
### Vamos a modificar esto para que sean listas de paises y asi poder analizarlos de una manera mas limpia
import numpy as np
df_copy = df.copy()
df_copy['country'] = df_copy['country'].apply(lambda row: row.split(',') if type(row) != float else row)
df_copy['country'] = df_copy['country'].apply(lambda row: [country.strip() for country in row] if type(row) == list else row)
```

```python
### Nuevamente, tenemos muchos generos que estan juntos en un string, estaria bueno estandarizar/homogeneizar esto en listas
import re
df_copy['listed_in'] = df_copy['listed_in'].apply(lambda row: re.split(',|&', row) if type(row) != float else row)
df_copy['listed_in'] = df_copy['listed_in'].apply(lambda row: [genre.strip() for genre in row] if type(row) == list else row)
```
